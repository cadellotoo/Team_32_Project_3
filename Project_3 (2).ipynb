```python
# Insert Caregiver Class
import random  # Import random module for shuffling

import random  # Import random module for shuffling

class Caregiver:
    def __init__(self, name, phone, email, pay_rate):
        self.name = name
        self.phone = phone
        self.email = email
        self.pay_rate = pay_rate
        self.hours_worked = 0
        self.availability = {}  # Format: {"7:00 AM - 1:00 PM": "available"}

    def update_availability(self, shift, status):
        self.availability[shift] = status

    def add_hours(self, hours):
        self.hours_worked += hours

    def get_contact_info(self):
        return f"{self.name}: {self.phone}, {self.email}"


class Shift:
    def __init__(self, shift_time):
        self.shift_time = shift_time
        self.caregiver = None

    def assign_caregiver(self, caregiver):
        self.caregiver = caregiver
        caregiver.add_hours(6)  # Each shift is 6 hours

    def clear_assignment(self):
        self.caregiver = None


class Schedule:
    def __init__(self):
        self.shifts = {}  # Format: {"YYYY-MM-DD": [Shift("7:00 AM - 1:00 PM"), ...]}
        self.caregivers = []

    def add_caregiver(self, caregiver):
        self.caregivers.append(caregiver)

    def generate_schedule(self):
        # Shuffle caregivers to randomize assignment
        for date, shift_list in self.shifts.items():
            for shift in shift_list:
                random.shuffle(self.caregivers)  # Randomize caregivers list
                for caregiver in self.caregivers:
                    if caregiver.availability.get(shift.shift_time) == 'preferred':
                        shift.assign_caregiver(caregiver)
                        break
                else:
                    # Assign 'available' caregiver if no 'preferred' found
                    random.shuffle(self.caregivers)
                    for caregiver in self.caregivers:
                        if caregiver.availability.get(shift.shift_time) == 'available':
                            shift.assign_caregiver(caregiver)
                            break

    def view_schedule(self):
        # Display the schedule
        unassigned_shifts = []
        for date, shift_list in self.shifts.items():
            print(f"{date}:")
            for shift in shift_list:
                caregiver_name = shift.caregiver.name if shift.caregiver else "Unassigned"
                print(f"  {shift.shift_time}: {caregiver_name}")
                if shift.caregiver is None:
                    unassigned_shifts.append((date, shift.shift_time))
        if unassigned_shifts:
            print("\nUnassigned Shifts:")
            for date, time in unassigned_shifts:
                print(f"  {date} - {time}")


# Example usage
schedule = Schedule()

# Adding caregivers
caregivers = [
    Caregiver("John", "123-456-7890", "John@example.com", 20),
    Caregiver("Jack", "987-654-3210", "jack@example.com", 12),
    Caregiver("Alex", "555-123-4567", "alex@example.com", 30),
    Caregiver("Mike", "444-222-1111", "mike@example.com", 19),
    Caregiver("Joe", "333-777-8888", "joe@example.com", 20),
    Caregiver("Josh", "888-999-0000", "josh@example.com", 21),
    Caregiver("Tom", "111-222-3333", "tom@example.com", 48),
    Caregiver("Daniel", "222-333-4444", "daniel@example.com", 20),
]

# Setting up caregivers' availability (Adjusted to include more 'available' caregivers)
caregivers[0].update_availability("7:00 AM - 1:00 PM", "preferred")
caregivers[0].update_availability("1:00 PM - 7:00 PM", "available")

caregivers[1].update_availability("7:00 AM - 1:00 PM", "available")
caregivers[1].update_availability("1:00 PM - 7:00 PM", "available")

caregivers[2].update_availability("7:00 AM - 1:00 PM", "available")
caregivers[2].update_availability("1:00 PM - 7:00 PM", "available")

caregivers[3].update_availability("7:00 AM - 1:00 PM", "available")
caregivers[3].update_availability("1:00 PM - 7:00 PM", "available")

caregivers[4].update_availability("7:00 AM - 1:00 PM", "available")
caregivers[4].update_availability("1:00 PM - 7:00 PM", "unavailable")

caregivers[5].update_availability("7:00 AM - 1:00 PM", "available")
caregivers[5].update_availability("1:00 PM - 7:00 PM", "available")

caregivers[6].update_availability("7:00 AM - 1:00 PM", "unavailable")
caregivers[6].update_availability("1:00 PM - 7:00 PM", "available")

caregivers[7].update_availability("7:00 AM - 1:00 PM", "preferred")
caregivers[7].update_availability("1:00 PM - 7:00 PM", "available")

# Adding caregivers to the schedule
for caregiver in caregivers:
    schedule.add_caregiver(caregiver)

# Adding shifts to the schedule
schedule.shifts["11-21-2024"] = [Shift("7:00 AM - 1:00 PM"), Shift("1:00 PM - 7:00 PM")]

# Generating and viewing the schedule
schedule.generate_schedule()
schedule.view_schedule()




#This is Madison Part
class Caregiver:
    def __init__(self, name, hours_worked):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = 20 #hourly rate is 20
        
    def weekly_pay(self):
        return self.hours_worked * self.hourly_rate #calculate weekly pay by multiplying hours worked by hourly rate(20)

def report(caregivers):
    weekly_total = 0
    monthly_total = 0  
    with open("pay_report.txt", "w") as report_file: #opening file to write report 
        report_file.write("Caregiver Pay Report\n")
        for caregiver in caregivers: #pay calculation loop
            gross = caregiver.weekly_pay()
            report_file.write(f" PAY REPORT: \n Name: {caregiver.name}   Hours Worked: {caregiver.hours_worked}   Weekly Pay: ${gross}\n") #return format
            weekly_total += gross
            monthly_total += gross * 4 #four weeks in a month (4 x weekly pay)  

        report_file.write(f"Total Weekly Pay: ${weekly_total}\n") #adding total to report 
        report_file.write(f"Total Monthly Pay (4 weeks): ${monthly_total}\n")


caregivers = [
    Caregiver("Jack", 12),
    Caregiver("Alex", 30),
    Caregiver("Tom", 48),
] #data input
report(caregivers) #call report function
with open("pay_report.txt", "r") as report_file:
    report_info = report_file.read() #read file
    print(report_info) #print output to verify 



#This is Ruhvir Part
import calendar
from datetime import datetime, timedelta

#created the list of caregivers and their times and stuff 
caregivers = [
    {"name": "John", "availability": {"7:00AM-1:00PM": True, "1:00PM-7:00PM": True}, "exceptions": ["11/28/24"]},
    {"name": "Jack", "availability": {"7:00AM-1:00PM": True, "1:00PM-7:00PM": True}, "exceptions": []},
    {"name": "Alex", "availability": {"7:00AM-1:00PM": False, "1:00PM-7:00PM": True}, "exceptions": ["11/12/24"]},
    {"name": "Mike", "availability": {"7:00AM-1:00PM": False, "1:00PM-7:00PM": False}, "exceptions": []},
    {"name": "Joe", "availability": {"7:00AM-1:00PM": True, "1:00PM-7:00PM": False}, "exceptions": ["11/27/24"]},
    {"name": "Josh", "availability": {"7:00AM-1:00PM": False, "1:00PM-7:00PM": False}, "exceptions": []},
    {"name": "Tom", "availability": {"7:00AM-1:00PM": False, "1:00PM-7:00PM": True}, "exceptions": []},
    {"name": "Daniel", "availability": {"7:00AM-1:00PM": False, "1:00PM-7:00PM": True}, "exceptions": []},
]


#creating the shifts based on the availability of the caregivers 
def createshifts(caregivers, year, month):
    schedule = {}
    days_in_month = calendar.monthrange(year, month)[1]

    for day in range(1, days_in_month + 1):        #iterates through the month 
        date = f"{month}/{day}/{year}"
        am_shift = []                    #creating then shifts that start at am and pm
        pm_shift = []
        for caregiver in caregivers:                            #loops through the if statements so that the available caregivers are called for those times
            if date not in caregiver["exceptions"]:
                if caregiver["availability"]["7:00AM-1:00PM"]:
                    am_shift.append(caregiver["name"])
                if caregiver["availability"]["1:00PM-7:00PM"]:
                    pm_shift.append(caregiver["name"])
        schedule[date] = {"7:00AM-1:00PM": am_shift, "1:00PM-7:00PM": pm_shift}
    
    return schedule          #return the schedule 

#create the schedule
month, year = 11, 2024 
schedule = createshifts(caregivers, year, month)

#create the html calendar  
class HTMLcal(calendar.HTMLCalendar):
    def formatday(self, day, weekday, schedule):
        if day == 0:
            return '<td class="noday">&nbsp;</td>'  #this is so that if day 0 which is a day not in calendar does not return anything and shows an empty cell
        date = f"{month}/{day}/{year}"
        shifts = schedule.get(date, {"7:00AM-1:00PM": [], "1:00PM-7:00PM": []}) 
        am_caregivers = ", ".join(shifts["7:00AM-1:00PM"])                  #retrieves the am and pm shifts 
        pm_caregivers = ", ".join(shifts["1:00PM-7:00PM"])
        return f'<td class="day"><b>{day}</b><br><b>7am-1pm:<b> {am_caregivers}<br><b>1pm-7pm:<b> {pm_caregivers}</td>' 
        #in the html I put the times that it will show and used <b> to bold it 

    def formatmonth(self, theyear, themonth, schedule, withyear=True): #formatting the month into html 
        v = []            #empty list that holds html strings 
        a = v.append          #appends items to the list 
        a('<table border="40" cellpadding="40" cellspacing="40" class="month">') #created big spaces using 40 to allow for all the text to fit into the days 
        a(self.formatmonthname(theyear, themonth, withyear=withyear))   #adds the formated month 
        a(self.formatweekheader())
        for week in self.monthdays2calendar(theyear, themonth):     #for loop that loops through the formatted month 
            a(self.formatweek(week, schedule))
        a('</table>')
        return ''.join(v)

    def formatweek(self, theweek, schedule):
        s = ''.join(self.formatday(d, wd, schedule) for d, wd in theweek)
        return f'<tr>{s}</tr>'

#create the HTML calendar
html_calendar = HTMLcal()
html_output = html_calendar.formatmonth(year, month, schedule)

#save to file
with open("care1.html", "w") as file:
    file.write(html_output)

print("created")
#I made all my code in jupiter notebook along with all my other teammates and then we added it into guthub 




#Calander 
import calendar
from datetime import datetime, timedelta

caregivers = [
    {"name": "John", "availability": {"AM": True, "PM": True}, "exceptions": ["11/28/24"]},
    {"name": "Jack", "availability": {"AM": True, "PM": False}, "exceptions": []},
    {"name": "Alex", "availability": {"AM": False, "PM": True}, "exceptions": ["11/12/24"]},
    {"name": "Mike", "availability": {"AM": False, "PM": False}, "exceptions": []},
    {"name": "Joe", "availability": {"AM": True, "PM": False}, "exceptions": ["11/27/24"]},
    {"name": "Josh", "availability": {"AM": False, "PM": False}, "exceptions": []},
    {"name": "Tom", "availability": {"AM": False, "PM": True}, "exceptions": []},
    {"name": "Daniel", "availability": {"AM": False, "PM": True}, "exceptions": []},
]

def createshifts(caregivers, year, month):
    schedule = {}
    days_in_month = calendar.monthrange(year, month)[1]

    for day in range(1, days_in_month + 1):
        date = f"{month}/{day}/{year}"
        am_shift = []
        pm_shift = []
        for caregiver in caregivers:
            if date not in caregiver["exceptions"]:
                if caregiver["availability"]["AM"]:
                    am_shift.append(caregiver["name"])
                if caregiver["availability"]["PM"]:
                    pm_shift.append(caregiver["name"])
        schedule[date] = {"AM": am_shift, "PM": pm_shift}
    
    return schedule

month, year = 11, 2024
schedule = createshifts(caregivers, year, month)

class HTMLCalendar(calendar.HTMLCalendar):
    def formatday(self, day, weekday, schedule):
        if day == 0:
            return '<td class="noday">&nbsp;</td>'  # Empty cell
        date = f"{year}-{month:02d}-{day:02d}"
        shifts = schedule.get(date, {"AM": [], "PM": []})
        am_caregivers = ", ".join(shifts["AM"])
        pm_caregivers = ", ".join(shifts["PM"])
        return f'<td class="day"><b>{day}</b><br>7am-1pm: {am_caregivers}<br>1pm-7pm: {pm_caregivers}</td>'

    def formatmonth(self, theyear, themonth, schedule, withyear=True):
        v = []
        a = v.append
        a('<table border="40" cellpadding="40" cellspacing="40" class="month">')
        a(self.formatmonthname(theyear, themonth, withyear=withyear))
        a(self.formatweekheader())
        for week in self.monthdays2calendar(theyear, themonth):
            a(self.formatweek(week, schedule))
        a('</table>')
        return ''.join(v)

    def formatweek(self, theweek, schedule):
        s = ''.join(self.formatday(d, wd, schedule) for d, wd in theweek)
        return f'<tr>{s}</tr>'

#create the HTML calendar
html_calendar = HTMLCalendar()
html_output = html_calendar.formatmonth(year, month, schedule)

# Save to file
with open("care.html", "w") as file:
    file.write(html_output)

print("created")

```
